# Data Preprocessing
[PCBA](https://s3-us-west-1.amazonaws.com/deepchem.io/datasets/molnet_publish/pcba.zip) -- Selected from PubChem BioAssay, consisting of measured biological activities of small molecules generated by high-throughput screening.

[MUV](https://s3-us-west-1.amazonaws.com/deepchem.io/datasets/molnet_publish/muv.zip) -- Subset of PubChem BioAssay by applying a refined nearest neighbor analysis, designed for validation of virtual screening techniques.

[HIV](https://s3-us-west-1.amazonaws.com/deepchem.io/datasets/molnet_publish/hiv.zip) -- Experimentally measured abilities to inhibit HIV replication.

[BACE](https://s3-us-west-1.amazonaws.com/deepchem.io/datasets/molnet_publish/bace.zip) -- Quantitative (IC50) and qualitative (binary label) binding results for a set of inhibitors of human Î²-secretase 1(BACE-1).

[CHEMBL](ftp://ftp.ebi.ac.uk/pub/databases/chembl/ChEMBLdb/latest/chembl_26.sdf.gz)  -- ext.
